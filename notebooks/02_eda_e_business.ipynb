# Datathon Passos Mágicos - Fase 5
## 02. Análise Exploratória e Respostas de Negócio (EDA)

**Objetivo:** Extrair insights estratégicos e responder às perguntas de negócio da Associação Passos Mágicos para apoiar a tomada de decisão.

**Padrão de Entrega:** Executivo / Diretoria

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings

warnings.filterwarnings('ignore')
sns.set_theme(style="whitegrid")
pd.set_option('display.max_columns', None)

CLEAN_DATA_PATH = '../local_data/dataset_limpo.csv'
print("Ambiente de análise configurado.")
```

### 1. Carga dos Dados
Base de dados limpa e padronizada na Fase 1.

```python
df = pd.read_csv(CLEAN_DATA_PATH, sep=';')
print(f"Dataset carregado: {df.shape[0]} alunos e {df.shape[1]} indicadores.")
```

### 2. Respostas de Negócio

#### 2.1 Adequação do Nível (IAN)
**Pergunta:** Qual o perfil de defasagem e como ele se comporta?

```python
plt.figure(figsize=(10, 4))
sns.histplot(data=df, x='IAN_2022', kde=True, color='teal')
plt.title('Perfil de Adequação ao Nível (IAN) - 2022')
plt.show()

print(f"Mediana IAN: {df['IAN_2022'].median()}")
print(f"% Alunos em Nível Crítico (IAN < 5): {(df['IAN_2022'] < 5).mean()*100:.2f}%")
```

**Análise Executiva (IAN):**
- A maioria dos alunos apresenta um IAN saudável (acima de 7.0).
- No entanto, identificamos que aproximadamente 20% do corpo discente está em nível crítico (IAN < 5), o que exige um plano de nivelamento acelerado para evitar o abandono.

#### 2.2 Desempenho Acadêmico (IDA)
**Pergunta:** O IDA está melhorando ou estagnado entre as fases?

```python
plt.figure(figsize=(12, 5))
sns.boxplot(data=df, x='FASE', y='IDA_2022', palette='viridis')
plt.title('Desempenho Acadêmico (IDA) por Fase')
plt.show()
```

**Análise Executiva (IDA):**
- Nota-se uma maior dispersão de notas nas fases iniciais (0 a 2).
- Nas fases finais (7 e 8), a mediana sobe e a dispersão diminui, indicando que os alunos que permanecem no programa atingem um padrão de excelência acadêmica mais homogêneo.

#### 2.3 Engajamento (IEG) vs Resultado (IDA)
**Pergunta:** O engajamento reflete diretamente no desempenho?

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='IEG_2022', y='IDA_2022', hue='PONTO_VIRADA_2022', alpha=0.6)
plt.title('Engajamento vs Desempenho')
plt.show()
```

**Análise Executiva (IEG):**
- O engajamento é o motor do desempenho: raramente um aluno com IEG baixo atinge IDA alto.
- O Ponto de Virada (pontos em laranja) concentra-se quase exclusivamente no quadrante superior direito (Alto Engajamento e Alto Desempenho).

#### 2.4 Autoavaliação (IAA) e Percepção
**Pergunta:** O aluno tem clareza sobre seu desempenho?

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(df['IAA_2022'], label='Autoavaliação (IAA)', fill=True)
sns.kdeplot(df['IDA_2022'], label='Nota Real (IDA)', fill=True)
plt.title('Autoavaliação vs Desempenho Real')
plt.legend()
plt.show()
```

**Análise Executiva (IAA):**
- Há um descolamento: a curva de autoavaliação (IAA) é mais otimista que a realidade (IDA).
- Isso sugere um espaço para trabalhar o amadurecimento e a autocrítica dos alunos, alinhando expectativas com resultados práticos.

#### 2.5 Aspectos Psicossociais (IPS)
**Pergunta:** O bem-estar psicológico afeta o INDE?

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='REC_PSICOLOGIA_2022', y='INDE_2022', palette='Set2')
plt.title('Impacto do Suporte Psicológico no Índice Global (INDE)')
plt.show()
```

**Análise Executiva (IPS):**
- Alunos classificados como "Sem limitações" psicossociais possuem INDE mediano significativamente superior.
- Aqueles que requerem avaliação apresentam maior volatilidade nas notas, confirmando que o suporte emocional é base para o sucesso acadêmico.

#### 2.6 Efetividade por Categoria (Pedra)
**Pergunta:** A evolução nas categorias reflete a melhoria dos indicadores?

```python
pedra_order = ['Quartzo', 'Ágata', 'Ametista', 'Topázio']
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='PEDRA_2022', y='INDE_2022', order=pedra_order, palette='coolwarm')
plt.title('Média do INDE por Categoria de Pedra')
plt.show()
```

**Análise Executiva (Efetividade):**
- O programa mostra-se extremamente efetivo: há um aumento degrau a degrau no INDE conforme o aluno avança de Quartzo para Topázio.
- A categoria **Topázio** é o objetivo aspiracional, mantendo médias de INDE próximas a 9.0.

### 3. Conclusão da EDA
- **Fator Crítico:** O engajamento (IEG) é o preditor mais forte de sucesso.
- **Foco de Intervenção:** Alunos em fases intermediárias com IAN < 5.
- **Oportunidade:** Alinhamento entre IAA e IDA para reduzir frustrações.
