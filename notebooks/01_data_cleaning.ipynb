# Datathon Passos Mágicos - Fase 5
## 01. Limpeza e Preparação de Dados

**Objetivo:** Consolidar a base de dados de 2024, padronizar nomenclaturas e preparar o dataset para análise exploratória.

**Padrão de Entrega:** Executivo / Diretoria

```python
import pandas as pd
import numpy as np
import os
import warnings

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)

DATA_PATH = '../local_data/DATATHON/'
print(f"Arquivos disponíveis: {os.listdir(DATA_PATH)}")
```

### 1. Carga dos Dados
Carregando a base principal consolidada em 2024.

```python
file_name = 'BASE DE DADOS PEDE 2024 - DATATHON.xlsx'
df_raw = pd.read_excel(os.path.join(DATA_PATH, file_name))

print(f"Base carregada com {df_raw.shape[0]} registros e {df_raw.shape[1]} colunas.")
```

### 2. Padronização e Renomeação
Mapeando as colunas conforme identificado na inspeção técnica para o padrão SNAKE_CASE profissional.

```python
map_cols = {
    'RA': 'RA',
    'Fase': 'FASE',
    'Turma': 'TURMA',
    'Nome': 'NOME',
    'Ano nasc': 'ANO_NASCIMENTO',
    'Idade 22': 'IDADE_2022',
    'Gênero': 'GENERO',
    'Ano ingresso': 'ANO_INGRESSO',
    'Instituição de ensino': 'INSTITUICAO_ENSINO',
    'Pedra 20': 'PEDRA_2020',
    'Pedra 21': 'PEDRA_2021',
    'Pedra 22': 'PEDRA_2022',
    'INDE 22': 'INDE_2022',
    'Cg': 'RANKING_GERAL_2022',
    'Cf': 'RANKING_FASE_2022',
    'Ct': 'RANKING_TURMA_2022',
    'Nº Av': 'QTD_AVALIACOES_2022',
    'IAA': 'IAA_2022',
    'IEG': 'IEG_2022',
    'IPS': 'IPS_2022',
    'IDA': 'IDA_2022',
    'Matem': 'NOTA_MAT_2022',
    'Portug': 'NOTA_PORT_2022',
    'Inglês': 'NOTA_ING_2022',
    'Indicado': 'INDICADO_BOLSA_2022',
    'Atingiu PV': 'PONTO_VIRADA_2022',
    'IPV': 'IPV_2022',
    'IAN': 'IAN_2022',
    'Fase ideal': 'FASE_IDEAL_2022',
    'Defas': 'DEFASAGEM_2022',
    'Rec Psicologia': 'REC_PSICOLOGIA_2022',
    'Destaque IEG': 'DESTAQUE_IEG_2022',
    'Destaque IDA': 'DESTAQUE_IDA_2022',
    'Destaque IPV': 'DESTAQUE_IPV_2022'
}

df_clean = df_raw.rename(columns=map_cols)
print("Colunas padronizadas com sucesso.")
```

### 3. Tratamento de Tipos e Nulos
Garantindo a integridade numérica para os cálculos de indicadores de performance.

```python
cols_to_numeric = [
    'INDE_2022', 'IAA_2022', 'IEG_2022', 'IPS_2022', 'IDA_2022', 
    'NOTA_MAT_2022', 'NOTA_PORT_2022', 'NOTA_ING_2022', 'IPV_2022', 'IAN_2022', 'DEFASAGEM_2022'
]

for col in cols_to_numeric:
    if col in df_clean.columns:
        df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')

# Preenchimento básico para evitar quebras em visualizações
df_clean['PEDRA_2022'] = df_clean['PEDRA_2022'].fillna('Não Avaliado')
df_clean['PONTO_VIRADA_2022'] = df_clean['PONTO_VIRADA_2022'].fillna('Não')

print("Tratamento de tipos concluído.")
```

### 4. Exportação Final
Salvando o dataset pronto para as fases de EDA e Machine Learning.

```python
OUTPUT_PATH = '../local_data/dataset_limpo.csv'
df_clean.to_csv(OUTPUT_PATH, index=False, sep=';', encoding='utf-8-sig')

print(f"\nSUCESSO: Fase 1 finalizada.\nArquivo gerado em: {OUTPUT_PATH}")
print(f"Total de Alunos processados: {len(df_clean)}")
```
